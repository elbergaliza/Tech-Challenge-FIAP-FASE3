```python
def investigate_dataframe(df):
    """
    Realiza uma investigação dos metadados de um DataFrame do Pandas.

    Args:
        df (pd.DataFrame): O DataFrame a ser investigado.

    Returns:
        None: Imprime as informações no console.
    """
    print("--- Informações do DataFrame ---")
    print(f"Número de Linhas: {df.shape[0]}")
    print(f"Número de Colunas: {df.shape[1]}")
    print("\n--- Primeiras 10 Linhas ---")
    print(df.head(10))
    print("\n--- Últimas 10 Linhas ---")
    print(df.tail(10))
    print("\n--- Tipos de Dados por Coluna ---")
    print(df.info())
    print("\n--- Estatísticas Descritivas ---")
    print(df.describe(include='all'))
    print("\n--- Valores Ausentes ---")
    print(df.isnull().sum())
    print("\n--- Valores Únicos por Coluna (Top 5) ---")
    for col in df.columns:
        unique_values = df[col].nunique()
        print(f"  {col}: {unique_values} valores únicos")
        if unique_values < 50: # Para evitar imprimir muitos valores para colunas com muitos únicos
            print(f"    Valores: {df[col].unique()[:5]}")
```

## Orquestração da arquitetura:  
1. **Representação de Dados**  
Realizar a tokenização e montar os embedings usando a biblioteca __Spacy__ em ambos os ambientes computacionais. Esta biblioteca é utilizada no ecossistema de aprendizado de máquina para realizar tarefas como o download de arquivos necessários para a geração de embeddings. Ela auxilia na simulação e visualização de processos internos dos modelos, como a tokenização, que consiste em quebrar uma frase em pedaços menores (tokens) para que o modelo possa processá-los matematicamente.  

2. **Otimização e Eficiência Computacional**
* Realizar a quantização do modelo.  
* Bibliotecas que deverão ser utilizadas:  
    * __Unsloth AI__: é uma ferramenta de código aberto que permite um treinamento muito mais rápido. Sua principal vantagem é o uso reduzido de memória e processamento, permitindo que o fine-tuning seja realizado mesmo em hardwares ou GPUs mais antigas e limitadas.  
    * __PEFT (Parameter-Efficient Fine-Tuning)__: É uma biblioteca voltada para o ajuste de modelos de forma computacionalmente eficiente, focando em alterar apenas uma pequena fração dos pesos do modelo original.   Ela é utilizada especificamente para aplicar a técnica LoRA (Low-Rank Adaptation), que evita o retreinamento de bilhões de parâmetros, focando apenas nos neurônios necessários para refinar o resultado da tarefa específica.  
    * __Transformers__: Baseada em uma arquitetura desenvolvida pelo Google, esta biblioteca utiliza mecanismos de atenção que permitem ao modelo processar todo o input de uma só vez, em vez de fazê-lo de forma sequencial. Essa abordagem facilita a paralelização eficiente do aprendizado e melhora a captura do sentido global do texto, sendo essencial para a compreensão de contextos em sequências longas.  
	* __Datasets__: Esta biblioteca é empregada para facilitar o carregamento e a formatação de conjuntos de dados. Ela permite que grandes volumes de informação sejam estruturados rapidamente em formatos que os modelos de linguagem consigam interpretar para o treinamento.  
3.  **Execução do FineTunning**  
    * Uso da biblioteca Unsloth AI, para reduzir drasticamente o consumo de VRAM e o tempo de processamento, garantindo que o treinamento seja performático e financeiramente sustentável.
    * Durante o Fine-tunning, salvar versões intermediarias do modelo
    * Salve o modelo e o tokenizador apos o termino do treinamento na pasta _pre-trained_
4.  **Teste do modelo Treinado**

    * Execute testes no modelo usando:
